````python
import sys
import io
import contextlib
import unittest
from langchain_core.messages import SystemMessage
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.prompts import HumanMessagePromptTemplate

# Setup the prompt templates
human_promt = HumanMessagePromptTemplate.from_template("{request}")
chat_prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a highly skilled data scientist with 20 years of experience. You specialize in writing clean, efficient, and error-free ML code. Generate code snippets without comments and ensure they are ready to execute."),
    human_promt
])
# Ensure reproducibility by setting a seed
model = Ollama(model="llama3", temperature=0)

# Function to generate code with LLM


def generate_code(request):
    formatted_request = chat_prompt.format_prompt(
        request=request).to_messages()
    response = model.invoke(formatted_request)
    generated_code = response
    return generated_code

# Function to clean and correct the code


def clean_and_correct_code(generated_code, csv_path):
    cleaned_code = generated_code.replace("```", "").strip()
    if "Here is the code snippet:" in cleaned_code:
        cleaned_code = cleaned_code.split(
            "Here is the code snippet:")[1].strip()
    if "python" in cleaned_code:
        cleaned_code = cleaned_code.split("python")[1].strip()
    corrected_code = cleaned_code.replace("{csv_path}", f"'{csv_path}'")
    return corrected_code

# Function to validate the corrected code


def validate_code(code):
    with contextlib.redirect_stdout(io.StringIO()) as f:
        try:
            exec(code)
            output = f.getvalue()
            return True, output
        except Exception as e:
            return False, f"Exception: {str(e)}\nCode:\n{code}"

# Function to request the LLM to fix the code


def request_fix_code(generated_code, error_message):
    fix_prompt = (
        "The following code generated an error during validation:\n"
        "```\n" + generated_code + "\n```\n"
        "Error message:\n"
        "```\n" + error_message + "\n```\n"
        "Please fix the code to resolve the error."
    )
    return generate_code(fix_prompt)


# Generate initial code
request = (
    "Write me a code for reading a CSV file, preprocessing the data (handling missing values, encoding categorical variables), "
    "splitting the data into training and test sets, training a decision tree classifier, and evaluating the model's performance. "
    "Use placeholders like {csv_path} for dynamic inputs."
)
generated_code = generate_code(request)
csv_path = "/Users/ilya/Desktop/GitHub_Repositories/HW_University/Data_Mining/datasets/Iris.csv"
corrected_code = clean_and_correct_code(generated_code, csv_path)

# Print cleaned and corrected code for debugging
print("Corrected code:")
print(corrected_code)
print("\nValidating code...")

# Write the corrected code to a temporary file
code_filename = "temp_generated_code.py"
with open(code_filename, "w") as file:
    file.write(corrected_code)

# Create a unittest test case to validate the generated code


class TestGeneratedCode(unittest.TestCase):
    def test_generated_code(self):
        with open(code_filename, "r") as file:
            code = file.read()
        exec(code)

# Function to run the unittest and return error messages


def run_unittest():
    suite = unittest.TestLoader().loadTestsFromTestCase(TestGeneratedCode)
    test_result = unittest.TextTestRunner().run(suite)
    return test_result


# Run the unittest
test_result = run_unittest()

if test_result.wasSuccessful():
    # If the test passes, save the validated code
    final_code_filename = "validated_generated_code.py"
    with open(final_code_filename, "w") as file:
        file.write(corrected_code)
    print(f"Generated code validated and saved to {final_code_filename}")
else:
    # If validation fails, capture the error and ask LLM to fix it
    error_message = str(test_result)
    fixed_code = request_fix_code(corrected_code, error_message)
    corrected_fixed_code = clean_and_correct_code(fixed_code, csv_path)

    # Write the fixed code to a temporary file for validation
    with open(code_filename, "w") as file:
        file.write(corrected_fixed_code)

    # Re-run the unittest for the fixed code
    test_result = run_unittest()

    if test_result.wasSuccessful():
        final_code_filename = "validated_fixed_code.py"
        with open(final_code_filename, "w") as file:
            file.write(corrected_fixed_code)
        print(f"Fixed code validated and saved to {final_code_filename}")
    else:
        print(f"Code validation failed again with errors:\n{test_result}")

# Cleanup: Remove the temporary file
os.remove(code_filename)
````
